```python
import numpy as np
import pandas as pd
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def softmax(b):
    exp = np.exp(b)
    s = np.sum(exp)
    return exp / s
```

```python
alpha = np.array([[1,1,2,-3,0,1,-3],
    [1,3,1,2,1,0,2],
    [1,2,2,2,2,2,1],
    [1,1,0,2,1,-2,2]], dtype=float)
beta = np.array([[1,1,2,-2,1],
       [1,1,-1,1,2],
       [1,3,1,-1,1]], dtype=float)
dbeta = None
dalpha = None
x = np.array([1,1,1,0,0,1,1])
y = 1
lr = 1
```

```python
pd.DataFrame(alpha)
```

```python
pd.DataFrame(beta)
```

```python
a = np.dot(x, alpha.T)
z = sigmoid(a)
z_ = np.append([1], z)
b = z_.dot(beta.T)
y_prime = softmax(b)

```

```python
db = y_prime
db[y] -= 1
dbeta = np.asmatrix(db).T.dot(np.asmatrix(z_))
dz = db.T.dot(beta[:, 1:]).T
da = dz * z * (1-z)
dalpha = np.asmatrix(da).T.dot(np.asmatrix(x))
```

```python
beta += - lr * dbeta
alpha += - lr * dalpha
```
